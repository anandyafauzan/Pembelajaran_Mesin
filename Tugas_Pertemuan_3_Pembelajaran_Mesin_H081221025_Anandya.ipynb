**2. CORRELATION & CAUSATION**

Korelasi Pearson dan Nilai P dari 'tax_revenue_country_country' dan 'total_tax_rate_country'

```python
correlation = df['tax_revenue_country_country'].corr(df['total_tax_rate_country'])
p_value = df['tax_revenue_country_country'].corr(df['total_tax_rate_country'], method='pearson')
print("Koefisien Korelasi Pearson:", correlation)
print("Nilai-P:", p_value)
```

Korelasi merupakan suatu pengukuran sejauh mana nilai salingketergantungan antar variabel. Causation merupakan hubungan
antara sebab dan akibat antara dua variabel

Nilai P diatas <0,001 sehingga ada bukti kuat bahwa korelasinya signifikan sedangkan pada korelasi pearson yang mengukur ketergantungan antara dua variabel yaotu tax revenue country dan total tax rate county.

5. **ANOVA**

```python
# grouping results
df_gptest = df[['category', 'countryOfCitizenship', 'finalWorth']]
grouped_test1 = df_gptest.groupby(['category', 'countryOfCitizenship'], as_index=False).mean()
print(grouped_test1)
```

Melihat apakah "category" mempengaruhi "final worth"

```python
from scipy import stats
```

```python
# ANOVA
grouped_test2 = df.groupby('category')
f_val, p_val = stats.f_oneway(grouped_test2.get_group('Automotive')['finalWorth'], grouped_test2.get_group('Telecom')['finalWorth'], grouped_test2.get_group('Energy')['finalWorth'])
print("ANOVA results: F =", f_val, ", P =", p_val)
```

Hasil ANOVA ini menunjukkan F-score tidak besar menunjukkan korelasi tidak begitu kuat dan nilai P hampir 0 menyiratkan signifikansi statistik yang hampir pasti
